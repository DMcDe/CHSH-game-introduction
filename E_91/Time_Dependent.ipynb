```python
import network_module_E91
from scipy.optimize import minimize
import numpy as np
```

```python
import time

distances = [1, 5, 10, 20]
results = []

for d in distances:
    g = network_module_E91.Game(distance=d,
                                attenuation=0.02,
                                guarantee_entanglement=True)

    start_time = time.perf_counter()
    _, _, S, _ = g.run_e91(num_rounds=2500)
    end_time = time.perf_counter()

    elapsed = end_time - start_time
    results.append((d, S, elapsed))

print("\n--- E91 Performance ---")
for d, S, t in results:
    print(f"Distance: {d:4} m → S = {S:.4f} | Runtime = {t:.2f} s")
```
